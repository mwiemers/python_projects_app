## Suspicious-Transaction Detector

Your task is to write Python code that scans a list of recent card transactions and automatically flags any that look suspicious.

A transaction should be flagged as suspicious if any of these conditions is true:
 - High-value: Amounts of £1500 or higher
 - Travel outlier: Travel category with an amount of £800 or higher
 - Electronics outlier: Electronics category with an amount of £1200 or higher

When you run the code cell it should read the provided CSV file, apply those rules, and print out every transaction that was flagged.
