### Coding Challenge 1 - WORDL game - simple version

We encourage you to start with the simple version of WORDLE in Python. Once you have completed this and feel ready to take on an extra challenge, move on to the advanced version.

Minimal guidance on what the wordle game should do:
- Select a word from the wordle_words.txt randomly
- Print 6 lines of placeholder words as in the example gif
- Prompt the user to enter a 5-letter word
- Display a coulored version of the entered word (letter in correct position: green, letter in wrong position: yellow)
- Terminate the game when the user entered the correct word or has run out of guesses


<details>
    <summary> Click here to reveal/hide more concrete tips and hints.</summary>

1. Read in the wordle_words.txt file as a list. You first have to read the file in and then split the string into a list based on linebreaks.

2. Select an element from the list randomly as the solution.

3. Create a variable for the maximum number of guesses.

4. Create a variable to keep track of how many guesses the user has guessed.

5. Create a list named `guesses` that initially consists of 6 placeholder words that should look as in the example, i.e. each placeholder should be the string: `"_ _ _ _ _ "`

6. Write a function that prints each of the six guesses in a separate line.

7. Start a while loop that:
    - Prints the guesses.
    - Ask the user to enter their guess.
    - Updates the guesses list to overwrite the n-th placeholder with the entered guess.
    - Stop the while loop if the guess is correct or the user ran out of guesses.

8. Write two separate functions that print a string in green and yellow respectively. 
    - A string preceded by the `GREEN` string will be printed in green.
    - A string preceded by the `YELLOW` string will be printed in yellow.
    - A string preceded by the `END` string will be printed in black.

9. Write a feedback function that uses the green and yellow function to return a coloured version of the word that the player entered.
    - Any character from the solution that is in the correct position should be in green.
    - Any character from the solution that is in the wrong position should be in yellow.
    - Any character that is not part of the solution should be in black.

10. Include the feedback function in the while loop to colour the users inputs to provide hints about which letters are in the correct or wrong position.

### Coding Challenge 2 - WORDL game - advanced version

After having finished the basic version of the WORDLE game, you can add a few features to created a more polished version:
* Clear the output after each time the user enters a guess
* Test the WORDLE game using the solution "tatoo" and enter the guess "ttttt". Only the first two t's should be coloured. Adjust your `feedback` function to make the function behave that way.
* Ask the player to enter the guess again if it contains non-letter characters or has less than or more than 5 characters.
* Add an intro text that welcomes the player to the WORDLE game and asks them to guess the 5-letter word. This intro text should always be displayed together with all previously entered guesses of the player.
* Print all letters of the guessed word in uppercase
* Print the solution if the player ran out of guesses
