## SoundFree - Data Cleaning

The data is currently split across the two files `song_info_1.csv` and `song_info_2.csv`. You need to load the two csv files, merge and clean them, and then save them as a new csv file.

We recommend to break down the problem of cleaning the song data into the following smaller tasks.

1. Load the `song_info_1.csv` file into Python as list of lists.

2. Load the `song_info_2.csv` file into Python.

3. Join the song data into a single variable named `song_info`.

4. Change the name of the data fields at the top of `song_info` to title, genre, mood, artist, duration, type, added.

5. Load the date values from the added_column.csv file. Ensure that you load the dates as a normal and not a nested collection. Check this by printing the first element.

6. The date values from the added_column.csv file have leading and trailing white space. You have to create a new version of the date values where white space has been removed. Carry out the corresponding steps manually for the first 3 date values. In the next task, you will have to automate this process.

7. Instead of repeating the same process for all date values manually, do this by cycling through all date values.

8. Print the first twenty elements of `dates_clean`, but do this by cycling through the elements of `dates_clean` based on the index number.

9. Replace the missing date values from song_info with the dates from dates_clean. Start by doing this manually one value at a time for the frist three values.

10. Instead of repeating the same process for all date values manually, do this by cycling through all date values based on index number.

11. You might have noticed that the type values, which are the 2nd to last value in each list of song_info, seem to only have missing values. Verify that all values are missing, but don't do this manually. Instead, use an efficient solution based on the techniques that you have learned about so far.

12. Remove the type values from song_info. Do this by cycling through the songs from song_info and remove the element of the song that corresponds to the type.

13. Create a new file named `'song_info_clean.csv'` and write the final version of `song_info` to the file.
